Plots the histogram for different inter spike intervals

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

plt.rc("font", size=23)
```

```python
with open("pickle_jar/spike_gs_31_100_percinh_0.0.pkl", "rb") as f:
    spike = pickle.load(f)
# with open(r"E:\master_thesis\again\modular_network\all_excitatory\pickled_files\spike_0.0_r_0.01_1.pkl", "rb") as f:
#     spike = pickle.load(f)

num_neurons = len(spike)
```

```python
spike_interval = [[] for _ in range(num_neurons)]
mean_isi = np.zeros(num_neurons)
std_isi = np.zeros(num_neurons)
coff_variation = np.zeros(num_neurons)

for i in range(num_neurons):
    for j in range(1, len(spike[i])):
        spike_interval[i].append(spike[i][j] - spike[i][j - 1])
```

```python
hist, bins = np.histogram(spike_interval[0], bins=10)
print(hist, bins)

hist_normalised = hist/hist.sum()
plt.figure(figsize=(10, 8))
plt.bar(bins[:-1], hist_normalised, width=bins[1] - bins[0])
plt.yticks(np.arange(0, 1.1, 0.1))

plt.xlabel("Inter spike interval (ms)")
plt.ylabel("PDF")
plt.show()
```
